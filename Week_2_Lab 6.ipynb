### **Introduction to SVM, Decision Trees, and Evaluation Metrics**

---

#### **Introduction to SVM (Support Vector Machines)**

**Concept of SVM:**
- SVM is a supervised machine learning algorithm that can be used for both classification and regression tasks, though it is primarily used for classification. It works by finding the optimal hyperplane that best separates the classes in the feature space.

**Key Concepts:**
- **Hyperplane:** A decision boundary that separates different classes. In two dimensions, this is a line; in three dimensions, it is a plane; and in higher dimensions, it is called a hyperplane.
- **Margin:** The distance between the hyperplane and the nearest data points from either class. SVM aims to maximize this margin to ensure robust classification.
- **Support Vectors:** The data points that are closest to the hyperplane. These points influence the position and orientation of the hyperplane.
- **Kernel Functions:** Functions that transform the data into a higher-dimensional space where a hyperplane can be used to separate the classes. Common kernels include linear, polynomial, and radial basis function (RBF).

---

#### **Introduction to Decision Trees**

**Concept of Decision Trees:**
- Decision Trees are a type of supervised learning algorithm that can be used for both classification and regression tasks. They work by splitting the data into subsets based on the value of input features, forming a tree-like structure.

**Key Concepts:**
- **Nodes:** Points in the tree where the data is split.
- **Branches:** The result of the split that leads to further nodes or leaves.
- **Leaves:** The terminal nodes that represent the final output or decision.
- **Splitting Criteria:** Methods used to decide where to split the data. Common criteria include Gini impurity and information gain (entropy).

---

#### **Introduction to Evaluation Metrics**

**Importance of Evaluation Metrics:**
- Evaluation metrics are essential to assess the performance of machine learning models. They help determine how well a model is making predictions and identifying areas where it may need improvement.

### 3. Visualizing the Decision Boundary

**Explanation:** Visualizing the decision boundary helps us understand how the SVM model separates the different classes. Here, we will visualize the decision boundary using the first two features of the Iris dataset.

**Code:**



````python
# Define a function to plot the decision boundary
def plot_decision_boundary(X, y, model, title='Decision Boundary'):
    # Create a mesh to plot the decision boundary
    h = .02  # Step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()
````


**Detailed Explanation:**
- **`def plot_decision_boundary(X, y, model, title='Decision Boundary'):`** Defines a function to plot the decision boundary of a given model.
- **`h = .02`:** Sets the step size in the mesh. A smaller value gives a finer resolution for the decision boundary.
- **`x_min, x_max`:** Determine the minimum and maximum values of the first feature.
- **`y_min, y_max`:** Determine the minimum and maximum values of the second feature.
- **`xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))`:** Creates a mesh grid from the minimum and maximum values of the features. This grid will be used to plot the decision boundary.
- **`Z = model.predict(np.c_[xx.ravel(), yy.ravel()])`:** Predicts the labels for each point in the mesh grid using the trained model. The `predict` method returns the predicted labels for each point.
- **`Z = Z.reshape(xx.shape)`:** Reshapes the predicted labels to match the shape of the mesh grid.
- **`plt.contourf(xx, yy, Z, alpha=0.8)`:** Fills the contour plot with different colors for different classes. The `alpha` parameter controls the transparency of the colors.
- **`plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')`:** Plots the original data points. The `c=y` parameter colors the points according to their true labels, and `edgecolors='k'` gives the points black edges.
- **`plt.title(title)`, `plt.xlabel('Feature 1')`, `plt.ylabel('Feature 2')`:** Adds title and labels to the plot.

---

**Code for Visualizing Decision Boundary on Training and Test Data:**

---

**1. Concept of Decision Trees**

**Explanation:** Decision Trees are a type of supervised learning algorithm that can be used for both classification and regression tasks. They work by recursively splitting the data into subsets based on the value of input features, forming a tree-like structure.

**Key Concepts:**
- **Nodes:** Points in the tree where the data is split based on a feature value.
- **Branches:** The connections between nodes, representing the outcome of the split.
- **Leaves:** The terminal nodes that represent the final decision or output.
- **Root:** The topmost node of the tree, where the first split occurs.

---

**2. Splitting Criteria: Gini Impurity and Entropy**

**Explanation:** Decision Trees use splitting criteria to decide where to split the data at each node. The two most common criteria are Gini Impurity and Entropy.

**1. Calculating Evaluation Metrics for SVM**

**Explanation:** We will calculate accuracy, precision, recall, and F1 Score for the SVM model using the test data.

**Code:**

````python
# Import necessary metrics from scikit-learn
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculate accuracy for SVM model
accuracy_svm = accuracy_score(y_test, y_pred_svm)

# Calculate precision for SVM model
precision_svm = precision_score(y_test, y_pred_svm, average='weighted')

# Calculate recall for SVM model
recall_svm = recall_score(y_test, y_pred_svm, average='weighted')

# Calculate F1 Score for SVM model
f1_svm = f1_score(y_test, y_pred_svm, average='weighted')

# Print the metrics for SVM model
print(f'SVM - Accuracy: {accuracy_svm:.2f}')
print(f'SVM - Precision: {precision_svm:.2f}')
print(f'SVM - Recall: {recall_svm:.2f}')
print(f'SVM - F1 Score: {f1_svm:.2f}')
````

**Detailed Explanation:**
- **`accuracy_score(y_test, y_pred_svm)`:** Calculates the accuracy of the SVM model by comparing the true labels (`y_test`) with the predicted labels (`y_pred_svm`).
- **`precision_score(y_test, y_pred_svm, average='weighted')`:** Calculates the weighted precision of the SVM model. The `average='weighted'` parameter accounts for label imbalance.
- **`recall_score(y_test, y_pred_svm, average='weighted')`:** Calculates the weighted recall of the SVM model.
- **`f1_score(y_test, y_pred_svm, average='weighted')`:** Calculates the weighted F1 Score of the SVM model.
- **`print(f'SVM - Accuracy: {accuracy_svm:.2f}')`:** Prints the accuracy of the SVM model, formatted to two decimal places.
- **`print(f'SVM - Precision: {precision_svm:.2f}')`:** Prints the precision of the SVM model, formatted to two decimal places.
- **`print(f'SVM - Recall: {recall_svm:.2f}')`:** Prints the recall of the SVM model, formatted to two decimal places.
- **`print(f'SVM - F1 Score: {f1_svm:.2f}')`:** Prints the F1 Score of the SVM model, formatted to two decimal places.

````python
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Binarize the output
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
n_classes = y_test_bin.shape[1]

# Train the OneVsRestClassifier with the SVM model
svm_model_ovr = OneVsRestClassifier(SVC(kernel='linear', probability=True))
svm_model_ovr.fit(X_train, y_train)

# Predict probabilities for the test set
y_score = svm_model_ovr.decision_function(X_test)

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot all ROC curves
plt.figure()
colors = ['blue', 'red', 'green']
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic to multi-class')
plt.legend(loc="lower right")
plt.show()
````
